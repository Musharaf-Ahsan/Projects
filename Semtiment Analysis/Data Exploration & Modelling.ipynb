# Preprocessing and Modelling

To build the corpus from the 'verified_reviews' we perform the following - <br>
1. Replace any non alphabet characters with a space
2. Covert to lower case and split into words
3. Iterate over the individual words and if it is not a stopword then add the stemmed form of the word to the corpus

```python
data = pd.read_csv("Senti.csv")
data.head()
```

- 0 = Positive
- 1 = Strong Positive
- 2 = Neutral
- 3 = Negative 
- 4 = Strong Positive

```python
import pandas as pd

# Load the data from a CSV file
data = pd.read_csv('Senti.csv')

# Define the mapping from sentiment labels to integers
sentiment_mapping = {
    'Positive': 0,
    'Strong Positive': 1,
    'Neutral': 2,
    'Negative': 3,
    'Strong Negative': 4  # Assuming 'Strong Negative' instead of the duplicated 'Strong Positive'
}

# Replace the sentiment labels with integers for the entire column
data['label'] = data['label'].map(sentiment_mapping)

# Save the modified DataFrame back to a CSV file
# df.to_csv('path_to_your_output_file.csv', index=False)

# Display the modified DataFrame
print(data)
```

```python
data['label'].value_counts()
```

```python
corpus = []
stemmer = PorterStemmer()
for i in range(0, data.shape[0]):
  review = re.sub('[^a-zA-Z]', ' ', data.iloc[i]['text'])
  review = review.lower().split()
  review = [stemmer.stem(word) for word in review if not word in STOPWORDS]
  review = ' '.join(review)
  corpus.append(review)
```

Using Count Vectorizer to create bag of words

```python
cv = CountVectorizer(max_features = 100)

#Storing independent and dependent variables in X and y
X = cv.fit_transform(corpus).toarray()
y = data['label'].values
```

```python
#Saving the Count Vectorizer
pickle.dump(cv, open('Models/countVectorizer.pkl', 'wb'))
```

Checking the shape of X and y

```python
print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

Splitting data into train and test set with 30% data with testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 15)

print(f"X train: {X_train.shape}")
print(f"y train: {y_train.shape}")
print(f"X test: {X_test.shape}")
print(f"y test: {y_test.shape}")
```

```python
scaler = MinMaxScaler(feature_range=(0, 4))

X_train_scl = scaler.fit_transform(X_train)
X_test_scl = scaler.transform(X_test)
```
